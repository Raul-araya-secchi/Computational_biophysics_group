```python
import py3Dmol


#Define PLDDT_BANDS and their color
PLDDT_BANDS = [(0, 50, '#FF7D45'),
               (50, 70, '#FFDB13'),
               (70, 90, '#65CBF3'),
               (90, 100, '#0053D6')]


with open("human_v3.pdb") as ifile:
    system = "".join([x for x in ifile])

view = py3Dmol.view(width=400, height=300)
view.addModelsAsFrames(system)

color_map = {i: bands[2] for i, bands in enumerate(PLDDT_BANDS)}

#view.setStyle({'model': -1}, {'cartoon': {'colorscheme': {'prop': 'b', 'map': color_map}}})
view.setStyle({'cartoon':{'arrows':True, 'tubes':True,'colorscheme': {'prop':'b','gradient': 'roygb','min':50,'max':90}}})
view.zoomTo()
view.show()
```
